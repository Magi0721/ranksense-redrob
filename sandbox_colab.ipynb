```python
from google.colab import files
print('Upload sample_candidates.json (or any small candidates.jsonl-style file).')
uploaded = files.upload()
```

```python
import json

# Convert sample_candidates.json (a JSON array) into the .jsonl format
# rank.py expects, if that's what was uploaded.
fname = list(uploaded.keys())[0]
with open(fname) as f:
    data = json.load(f)

with open('sample_candidates.jsonl', 'w') as f:
    for c in data:
        f.write(json.dumps(c) + '\n')

print(f'Wrote {len(data)} candidates to sample_candidates.jsonl')
```

```python
!python src/rank.py --candidates sample_candidates.jsonl --out demo_submission.csv --xlsx-out demo_submission.xlsx
```
